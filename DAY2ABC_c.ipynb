```python
import cv2
import requests
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/image.png') # Read the image
cv2_imshow(image) # Display the image
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
resized_image = cv2.resize(image, (300, 200))
cv2_imshow(resized_image)
```

```python
flipped_ver = cv2.flip(image, 0)
cv2_imshow(flipped_ver)
```

```python
flipped_hor = cv2.flip(image, 1)
cv2_imshow(flipped_hor)
```

```python
edges = cv2.Canny(gray_image, 200, 300)
cv2_imshow(edges)
```

```python
black_image = np.zeros((500, 800, 3))
cv2_imshow(black_image)
```

```python
org = (80, 250)
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 2
color = (255, 0, 0)
thickness = 2
text = 'Welcome to OpenCV!'
black_image = cv2.putText(black_image, text, org, font, font_scale, color, thickness)
cv2_imshow(black_image)
```

```python
image_url = "https://img.freepik.com/free-vector/thank-you-lettering_1262-6963.jpg"
```

```python
response = requests.get(image_url)
image_data = np.asarray(bytearray(response.content), dtype="uint8")
```

```python
image = cv2.imdecode(image_data, cv2.IMREAD_COLOR)
```

```python
if image is None:
  raise ValueError("Failed to load image from the URL. Please check the URL.")
```

```python
cv2_imshow(image)
```
